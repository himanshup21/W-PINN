```python
ntk_wpinn = np.loadtxt('WPINN_ntk.csv', delimiter=',')
ntk_pinn = np.loadtxt('PINN_ntk.csv', delimiter=',')
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})

itr = [0, 1000, 10000, 50000]
colors = ['blue', 'green', 'red', 'purple']

plt.figure(figsize=(6, 5))

lines_wpinn = []
lines_pinn = []

for i in range(len(itr)):
    line_wpinn, = plt.plot(ntk_wpinn[i], '-', color=colors[i], label=f'itr={itr[i]}')
    line_pinn, = plt.plot(ntk_pinn[i], '--', color=colors[i])
    if i == 0:
        lines_wpinn.append(line_wpinn)
        lines_pinn.append(line_pinn)

plt.yscale('log')
plt.xlabel('Index', size=15)
plt.ylabel('NTK Eigenvalues', size=15)
plt.title('NTK Eigenvalues comparison', size=17)
plt.xticks(size=12)
plt.yticks(size=12)
plt.grid(True, which="both", ls="--")

plt.tight_layout()


color_lines = [plt.Line2D([0], [0], color=colors[i], lw=2) for i in range(len(itr))]
legend1 = plt.legend(color_lines, [f'itr={it}' for it in itr], title='Iterations', loc = (0.7,0.3), fontsize=12)

plt.gca().add_artist(legend1) 


style_lines = [
    plt.Line2D([0], [0], color='black', lw=2, linestyle='-'),
    plt.Line2D([0], [0], color='black', lw=2, linestyle='--')
]
legend2 = plt.legend(style_lines, ['WPINN', 'PINN'], title='Model Type', loc = (0.7,0.61), fontsize=12)

plt.savefig('ntk_comparison.jpg', bbox_inches='tight')
plt.savefig('NTK.eps', format='eps', dpi=300, bbox_inches='tight')
plt.show()
```
